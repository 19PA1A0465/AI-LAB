```
import pandas as pd # for handling data in the form of tables
import numpy as np # For handling matrix
from sklearn.model_selection import train_test_split # For spliting the data into train and test
from sklearn.linear_model import LinearRegression # for using the model
from sklearn import metrics
# import matplotlib.pyplot as plt # for plotting
```

```
ipl=pd.read_csv("students_placement_data.csv")
ipl
```

```
ipl.describe()# to study  all stastics summry
```

```
ipl.describe(include="all")# to study  all stastics summary
```

```
ipl
```

```
ipl.plot(x='SSC Percentage',y='inter_Diploma_percentage',style='*')# to how they are corelated
```

```
#converting a  series into numpy array
x=ipl['SSC Percentage'].values.reshape(-1,1)
x
y=ipl['inter_Diploma_percentage'].values.reshape(-1,1)
y
```

```
 #Now divide the data into train and test. Here we are keeping 40% as our test set
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.4)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)




```

```
# Now apply linear regression model on training dataset
model=LinearRegression()
model.fit(X_train,Y_train)
```

```
# to find the intercept of our model we use  intercept_
print("intercept=",model.intercept_) 
```

```
# To find  slope of our model we use coef_
print("slope=",model.coef_)
```

```
#a=predict interpercentage
a=model.predict(X_test) 
a
```

```
# b=actual prediction
b=pd.DataFrame({'predicted':a.flatten(), 'Actual':Y_test.flatten()})
b
```

```
import matplotlib.pyplot as plt
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_test,a)
```

```
print('R2 error:',metrics.r2_score(Y_test, a))
```
